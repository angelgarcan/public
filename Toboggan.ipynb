```python
with open("map_forest.txt", 'r', encoding='utf-8') as f:
    forest=[list(line.replace('\n','')) for line in f]

h=len(forest)
w=len(forest[0])

print(f">>>> FOREST (height={h}, width={w}) <<<<\n")
for line in forest:
    print(''.join(line))
```

```python
row=0
col=0
trip=""
XOs=""
while row<h:
    trip+=forest[row][col]
    XOs+='X' if trip[-1]=='#' else 'O'
    print(f"({row},{col})::({trip[-1]},{XOs[-1]})",end=' -> ')
    
    row+=1
    col+=3
    if col>=w: col=col%w

trees_count=XOs.count('X')

print("\n==============================")
print(f"TRIP :\n{trip}")
print(f"XOs :\n{XOs}")
print(f"Number of trees: {trees_count}")
```
